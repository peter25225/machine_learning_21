### Отобразим данные

```python
plt.scatter(data['x'], data['y'], 0.5)
plt.figure
```

Для кластеризации приходится брать только часть данных - на большей выборке "подвисает" алгоритм AgglomerativeClustering

```python
X = data.drop('comment_class', axis=1)  
y = data['comment_class']

X_sample, X_test, y_sample, y_test = train_test_split(X, y, test_size=0.7, random_state = 2020, stratify=y)
```

#### Нормируем данные для последующей кластеризации

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X_sample)
```

#### Оперделим оптимальное число кластеров KMeans

```python
k_inertia = []
ks = range(1,15)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled)
    k_inertia.append(clf_kmeans.inertia_)
    
plt.plot(ks, k_inertia)
```

```python
diff = np.diff(k_inertia)
plt.plot(ks[1:], diff)
```

```python
diff_r = diff[1:] / diff[:-1]
plt.plot(ks[1:-1], diff_r)
```

```python
k_opt = ks[np.argmin(diff_r)+1]
k_opt
```

Несмотря на то, что по расчету оптимальным является 3 кластера, возьмем 8 кластеров - это более соответствует числу районов в нашем понимании

```python
algorithms = [
# Исключаем алгорит affinityPropagation - он по какой-то причине не срабатывает
#     AffinityPropagation(),
     KMeans(n_clusters=8),
     AgglomerativeClustering(n_clusters=8),
     DBSCAN(eps=0.1, min_samples=8)
]
```

```python
for algorithm in algorithms:
    clusters = algorithm.fit_predict(X_scaled)
    plt.scatter(X_scaled[:, 0],X_scaled[:, 1], 0.75, c=clusters)
    plt.title(algorithm.__class__.__name__)
    if algorithm.__class__.__name__ in {'KMeans', 'AffinityPropagation'}:
        centers = algorithm.cluster_centers_
        plt.scatter(centers[:, 0], centers[:, 1], s=50, marker='x', color='black')
    plt.show()
```

#### Повторим кластеризацию отдельно для позитивных и негативных ответов

```python
data = pd.read_csv('geo.csv')
del data['Id']
data
```

```python
negative = data[data['comment_class'] == -1]
positive = data[data['comment_class'] == 1]

X = negative.drop('comment_class', axis=1)  
y = negative['comment_class']

X_sample_negative, X_test, y_sample, y_test = train_test_split(X, y, test_size=0.7, random_state = 2020, stratify=y)


X = positive.drop('comment_class', axis=1)  
y = positive['comment_class']

X_sample_positive, X_test, y_sample, y_test = train_test_split(X, y, test_size=0.7, random_state = 2020, stratify=y)
```

```python
plt.scatter(X_sample_positive['x'], X_sample_positive['y'], 0.5)
plt.figure
```

```python
plt.scatter(X_sample_negative['x'], X_sample_negative['y'], 0.5)
plt.figure
```

Заметно, что негативные комментарии сконцентрированы больше в центральных районах

```python
X_scaled_positive = scaler.fit_transform(X_sample_positive)
X_scaled_negative = scaler.fit_transform(X_sample_negative)
```

```python
algorithms = [
#      AffinityPropagation(),
     KMeans(n_clusters=8),
     AgglomerativeClustering(n_clusters=8),
     DBSCAN(eps=0.1, min_samples=8)
]
```
